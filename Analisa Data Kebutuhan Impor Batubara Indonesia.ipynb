Deskripsi Data
Data ini berisi mengenai jumlah produksi batubara tahun 2003 - 2010 dan jumlah konsumsi domestik batu bara tahun 2003 -2010. Dalam dataset ini berisikan variabel:

1. tipe_entitas : Tipe Entitas
2. nama_entitas : Nama Entitas
3. tahun: Tahun
4. nilai: Nilai


# Data Produksi Batubara 2003-2010

```python
data_produksi = pd.read_csv('processed-produksi-batubara-2003-2010.csv', delimiter = ',', encoding = "utf-8")
```

```python
data_produksi
```

```python
data_produksi.count()
```

```python
data_produksi.columns
```

```python
data_produksi.tipe_entitas.count()
```

```python
s = data_produksi.nilai.sum()
mi = data_produksi.nilai.min()
ma = data_produksi.nilai.max()

print("Total Produksi Batubara Tahun 2003 - 2010 adalah", s)
print("Total Nilai Produksi Terendah adalah",mi)
print("Total Nilai Produksi Tertinggi adalah", ma)
```

```python
data_produksi.groupby('tahun').mean()
```

# Visualisasi Sederhana Data Produksi Batubara 2003 - 2010

```python
ax = data_produksi.groupby('tahun').mean()['nilai']
ax = ax.plot(kind='bar', title ="Grafik Produksi Batu Bara 2003 - 2010", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Nilai", fontsize=12)
```
